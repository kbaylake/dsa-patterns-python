```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        def backtrack(candidates,i,t,ans):
            if i>=len(candidates) or t<0:
                return []
            if t==0:
                return [ans[:]]
            ans.append(candidates[i])
            include=backtrack(candidates[i:],i,t-candidates[i],ans)
            ans.pop()
            exclude=backtrack(candidates[i+1:],i+1,t,ans)
            return include+exclude
        return backtrack(candidates,0,target,[])x
```
