```python
class Node:
    def __init__(self,data,left,right):
        self.data=data
        self.left=left
        self.right=right
    def setData(self,value):
        self.data=value
    def setLeft(self,leftPointer):
        self.left=leftPointer
    def setRight(self,rightPointer):
        self.right=rightPointer
    def getData(self):
        return self.data
    def getLeft(self):
        return self.left
    def getRight(self):
        return self.right
```

```python
def searchTree(root,target):
    def search(parent,target):
        if not parent: #O(1)
            return -1
        val=parent.getData() #O(1)
        if val==target: #O(1)
            return val
        left=parent.getLeft()
        leftSearch=search(left,target)
        right=parent.getRight()
        rightSearch=search(right,target)
        if leftSearch!=-1:
            return leftSearch
        else:
            return rightSearch
    return search(root,target) 
```

```python
class BinarySearchTree:
    def __init__(self):
        self.root=None
    def addRoot(self):
        rootVal=int(input('Enter the root value'))
        root=Node(rootVal,None,None)
        self.root=root
        print(f'Root value set to {rootVal}')
    def addVal(self): 
        if not self.root:
            print('Creating root')
            self.addRoot()
            return
        val=int(input('Enter Value to add'))
        newNode=Node(val,None,None)
        self.add(self.root,val,newNode) 
        return
    def add(self,parent,val,newNode):
        if val>parent.getData():
            if not parent.getRight():
                parent.setRight(newNode) #O(1)
                print(f'Added {val} to the right of {parent.getData()}')
                return
            self.add(parent.getRight(),val,newNode)
        else:
            if not parent.getLeft():
                parent.setLeft(newNode) #O(1)
                print(f'Added {val} to the left of {parent.getData()}')
                return
            self.add(parent.getLeft(),val,newNode)
    def getRoot(self):
        return self.root
```

```python
def searchBSTree(root,target):
    def search(parent,target):
        if not parent: #O(1)
            return False
        val=parent.getData() #O(1)
        if val==target: #O(1)
            return True
        left=parent.getLeft()
        leftSearch=search(left,target)
        right=parent.getRight()
        rightSearch=search(right,target)
        if leftSearch!=-1:
            return leftSearch
        else:
            return rightSearch
    return search(root,target) 
```

```python
def search(parent,target):
    if not parent:
        return False
    val=parent.getData() #O(n)
    if target==val:
        return True
    if target>val:
        right=parent.getRight()
        return search(right,target)
    else:
        left=parent.getLeft()
        return search(left,target)
```

This algo halves the tree at each comparison by only considering the left or right subarray thus O(log(n)) <br>
The difference between Binary Tree and BST is that the searching for BST is log(n) vs O(n)

```python
def getHeight(parent,curr):
    if not parent:
        return curr
    left=getHeight(parent.getLeft(),curr+1)    
    right=getHeight(parent.getRight(),curr+1)
    return max(left,right)
```

Given a BINARY TREE you have to make a function which returns True if the Binary Tree is a Binary Search Tree the time complexity should be O(n) max without finding the inorder

98. Validate Binary Search Tree
https://leetcode.com/problems/validate-binary-search-tree/description/

```python
import math
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True # An empty tree can be a BST
        def backtrack(parent,isTrue,min,max):
            if not parent:
                return isTrue
            val=parent.val
            if val>min and val<max:
                left=backtrack(parent.left,True,min,val)
                right=backtrack(parent.right,True,val,max)
                return left and right 
            else:
                return False
        return backtrack(root,False,-math.inf,math.inf) 
```

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        m,n=len(matrix),len(matrix[0])
        low=0
        high=m*n-1
        while low<=high:
            mid=(low+high)//2
            row=mid//n
            col=mid%n
            val=matrix[row][col]
            if val==target:
                return True
            if val<target:
                low=mid+1
            else:
                high=mid-1
        return False
```

```python
def bfs(root):
    if not root:
        return []
    ans=[]
    queue=[]
    queue.append(root)
    ans.append(root.getData())
    while queue:
        curr=queue[0]
        queue.pop(0)
        if curr.getLeft():
            ans.append(curr.getLeft().getData())
            queue.append(curr.getLeft())
        if curr.getRight():
            ans.append(curr.getRight().getData())
            queue.append(curr.getRight())
    return ans
```
