Example 1: You are given a string s and an integer k. Find the length of the longest substring that contains at most k distinct characters.

For example, given s = "eceba" and k = 2, return 3. The longest substring with at most 2 distinct characters is "ece".

```python
from collections import defaultdict

def find_longest_substring(s, k):
    ans=left=right=0
    dc={}
    while right<len(s):
        dc[s[right]]+=1
        while len(dc)>k:
            dc[s[left]]-=1
            if dc[s[left]]==0:
                del dc[s[left]]
            left+=1
        ans=max(ans,right-left+1)
        right+=1
    return ans
    
```

Example 2: 2248. Intersection of Multiple Arrays

Given a 2D array nums that contains n arrays of distinct integers, return a sorted array containing all the numbers that appear in all n arrays.

For example, given nums = [[3,1,2,4,5],[1,2,3,4],[3,4,5,6]], return [3, 4]. 3 and 4 are the only numbers that are in all arrays.

```python
from collections import defaultdict
def intersection(nums):
    dc=defaultdict(int)
    for i in nums:
        for j in i:
            dc[j]+=1
    op=[i for i in dc if dc[i]==len(nums)]
    op.sort()
    return op
```

Example 3: 1941. Check if All Characters Have Equal Number of Occurrences

Given a string s, determine if all characters have the same frequency.

For example, given s = "abacbc", return true, because all characters appear twice. Given s = "aaabb", return false. "a" appears 3 times, "b" appears 2 times. 3 != 2.

```python
def areOccurrencesEqual(self, s: str) -> bool:
    dc=defaultdict(int)
    for i in s:
        dc[i]+=1
    return len(set(dc.values()))==1
```


Example 4: 560. Subarray Sum Equals K

Given an integer array nums and an integer k, find the number of subarrays whose sum is equal to k.

```python
def subarraySum(nums, k: int) -> int:
    dc=defaultdict(int)
    dc[0]=1
    curr=ans=0
    for i in nums:
        curr+=i
        if curr-k in dc:
            ans+=dc[curr-k]
        dc[curr]+=1
    return ans
```

Given an integer array nums, return the largest integer that only occurs once. If no integer occurs once, return -1.

 

Example 1:

Input: nums = [5,7,3,9,4,9,8,3,1]
Output: 8
Explanation: The maximum integer in the array is 9 but it is repeated. The number 8 occurs only once, so it is the answer.
Example 2:

Input: nums = [9,9,8,8]
Output: -1
Explanation: There is no number that occurs only once.
 

Constraints:

1 <= nums.length <= 2000
0 <= nums[i] <= 1000

```python
from collections import defaultdict
def largestUniqueNumber(nums) -> int:
    ans=-1
    dc=defaultdict(int)
    for num in nums:
        dc[num]+=1
    for i in dc:
        if dc[i]==1:
            ans=max(ans,i)
    return ans
```

Given a string text, you want to use the characters of text to form as many instances of the word "balloon" as possible.

You can use each character in text at most once. Return the maximum number of instances that can be formed.

```python
def maxNumberOfBalloons(text: str) -> int:
    textC=Counter(text)
    balC=Counter('balloon')
    ans=[]
    for char in balC:
        ans.append(textC[char]/balC[char])
    return int(min(ans))
```

Example 1: 49. Group Anagrams

Given an array of strings strs, group the anagrams together.

For example, given strs = ["eat","tea","tan","ate","nat","bat"], return [["bat"],["nat","tan"],["ate","eat","tea"]].

```python
def groupAnagrams(strs):
    dc=defaultdict(list)
    for i in strs:
        key=''.join(sorted(i))
        dc[key].append(i)
    return dc.values()
```

Example 2: 2260. Minimum Consecutive Cards to Pick Up

Given an integer array cards, find the length of the shortest subarray that contains at least one duplicate. If the array has no duplicates, return -1.

```python
def minimumCardPickup(cards) -> int:
    dc={}
    ans=float(inf)
    for i in range(len(cards)):
        if cards[i] in dc:
            ans=min(ans,i-dc[cards[i]]+1)
        dc[cards[i]]=i
    return ans if ans < float("inf") else -1
```

You are given a 0-indexed array nums consisting of positive integers. You can choose two indices i and j, such that i != j, and the sum of digits of the number nums[i] is equal to that of nums[j].

Return the maximum value of nums[i] + nums[j] that you can obtain over all possible indices i and j that satisfy the conditions. If no such pair of indices exists, return -1.

 



```python
def maximumSum(nums) -> int:
    def digitSum(num):
        sum=0
        while num:
            sum+=num%10
            num//=10
        return sum
    dc=defaultdict(int)
    ans=-1
    for num in nums:
        key=digitSum(num)
        if key in dc:
            ans=max(ans,dc[key]+num)
        dc[key]=max(dc[key],num)
    
    return ans
```

Given a 0-indexed n x n integer matrix grid, return the number of pairs (ri, cj) such that row ri and column cj are equal.

A row and column pair is considered equal if they contain the same elements in the same order (i.e., an equal array).

 

```python
def equalPairs(grid) -> int:
    n=len(grid)
    row=[tuple(i) for i in grid]
    col=[tuple(grid[i][j] for i in range(n)) for j in range(n)]
    ans=0
    dc_row=defaultdict(int)
    dc_col=defaultdict(int)
    for i in range(n):
        dc_row[row[i]]+=1
        dc_col[col[i]]+=1
    for i in dc_row:
        if i in dc_col:
            ans+=dc_col[i]*dc_row[i]
    return ans
```

You're given strings jewels representing the types of stones that are jewels, and stones representing the stones you have. Each character in stones is a type of stone you have. You want to know how many of the stones you have are also jewels.

Letters are case sensitive, so "a" is considered a different type of stone from "A".

```python
def numJewelsInStones(jewels: str, stones: str) -> int:
    jewelsCount=Counter(jewels)
    stonesCount=Counter(stones)
    ans=0
    for i in stonesCount:
        if i in jewelsCount:
            ans+=stonesCount[i]
    return ans
```
