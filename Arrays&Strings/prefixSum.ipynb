Example 1: Given an integer array nums, an array queries where queries[i] = [x, y] and an integer limit, return a boolean array that represents the answer to each query. A query is true if the sum of the subarray from x to y is less than limit, or false otherwise.

For example, given nums = [1, 6, 3, 2, 7, 2], queries = [[0, 3], [2, 5], [2, 4]], and limit = 13, the answer is [true, false, true]. For each query, the subarray sums are [12, 14, 12].

```python
def ansQueries(nums,queries,limit):
    prefixs=[nums[0]]
    for i in range(1,len(nums)):
        prefixs.append(nums[i]+prefixs[-1])
    ans=[prefixs[y]-prefixs[x]+nums[x]<=limit for x,y in queries]
    return ans
    
```

```python
nums = [1, 6, 3, 2, 7, 2]
queries = [[0, 3], [2, 5], [2, 4]]
limit = 13
ansQueries(nums,queries,limit)
```

Example 2: 2270. Number of Ways to Split Array

Given an integer array nums, find the number of ways to split the array into two parts so that the first section has a sum greater than or equal to the sum of the second section. The second section should have at least one number.

```python
def pre(nums):
    prefixes=[nums[0]]
    for i in range(1,len(nums)):
        prefixes.append(nums[i]+prefixes[-1])
    return prefixes
pre([2,3,1,0])
```

```python
def split(nums):
    prefixes=[nums[0]]
    for i in range(1,len(nums)):
        prefixes.append(nums[i]+prefixes[-1])
    left=0
    right=len(prefixes)-1
    ans=0
    while left<right:
        if prefixes[left]>=prefixes[right]-prefixes[left]:
            ans+=1
        left+=1
    return ans
        
```

```python
split([10,4,-8,7])
```

```python
def splitWithoutTotal(nums):
    total=sum(nums)
    left_section=ans=0
    for i in range(len(nums)-1):
        left_section+=nums[i]
        right_section=total-left_section
        if left_section>=right_section:
            ans+=1
    return ans
        
```

```python
splitWithoutTotal([2,3,1,0])
splitWithoutTotal([10,4,-8,7])
```

Given an array of integers nums, you start with an initial positive value startValue.

In each iteration, you calculate the step by step sum of startValue plus elements in nums (from left to right).

Return the minimum positive value of startValue such that the step by step sum is never less than 1.

```python
def runningSum(msv,nums):
    prefixes=[msv+nums[0]]
    for i in range(1,len(nums)):
        curr=nums[i]+prefixes[-1]
        prefixes.append(curr)
    for i in prefixes:
        if i<1:
            return False
    return True
def minStartValue(nums):
    msv=1
    while runningSum(msv,nums)==False:
        msv+=1
    return msv
        
```

```python
def getAvg(nums,k):
    prefixes=[nums[0]]
    for i in range(1,len(nums)):
        prefixes.append(nums[i]+prefixes[-1])
    ans=[]
    n=len(nums)
    for i in range(n):
        if i-k>=0 and i+k<n:
            ans.append((prefixes[i+k]-prefixes[i-k]+nums[i-k])//(2*k+1))
        else:
            ans.append(-1)
    return ans
```

```python
def reverseWord(word):
    word_ls=[i for i in word]
    left=0
    right=len(word_ls)-1
    while left<right:
        word_ls[left]=word[right]
        word_ls[right]=word[left]
        left+=1
        right-=1
    return ''.join(word_ls)
def reverseStr(s):
    s=s.split()
    return ' '.join([reverseWord(word) for word in s])
        
```

```python
def minSub(nums,target):
    nums.sort(reverse=True)
    left=0
    right=0
    sum=0
    for right in range(len(nums)):
        sum+=nums[right]
        if sum<target:
            right+=1
        else:
            sum-=nums[left]
            left+=1
        return right-left+1
```
